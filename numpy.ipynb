```python
# 1. A matemática da IA (Numpy)
import numpy as np

print("--- O PROBLEMA DO PYTHON PURO ---")
precos_lista = [100000, 200000, 300000]

# O Python puro não faz a conta, ele apenas duplica os elementos da lista!
print("Lista vezes 2:", precos_lista * 2)

print("\n--- A SOLUÇÃO COM NUMPY ---")
#Transforma a lista em um 'Array' (veor matemático)
precos_numpy = np.array([100000, 200000, 300000])

# O Numpy percebe instantaneamente que queremos multiplicar cada número!
precos_atualizados = precos_numpy * 2
print("Array vezes 2:", precos_atualizados)

print("\n--- COMO A IA VÊ O MUNDO (Matrizes) ---")
# Na IA tudo são números (Matrizes)
matriz_dados = np.array([
    [1,2,3],
    [4,5,6]
])
print(matriz_dados)
```
