```python
# Conhecendo o Pandas
import pandas as pd

# 1. Criando um dicionário
dados_exemplos = {
    "Nome": ["Ana", "Bruno", "Carlos"],
    "Idade": [25, 30, 22],
    "Profissão": ["Engenheira", "Motorista", "Estudante"]
}

# 2. Tranformando o dicionário em uma tabela (Dataframe)
tabela = pd.DataFrame(dados_exemplos)

print("Olhem como o pandas organiza os dados lindamente:")
display(tabela)

3 #. Brincando com a tabela (Separando apenas a coluna Idade)
idades = tabela["Idade"]
print("\nApenas as coluna de idades:")
print(idades)
```
