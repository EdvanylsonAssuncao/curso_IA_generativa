```python
# Dados para a dinâmica
import pandas as pd

# Histórico dos meses anteriores
dados_marketing = {
    'Investimento_Anuncios': [1000, 2000, 3000, 4000, 5000],
    'Faturamento': [15000, 25000, 36000, 48000, 60000]
}

tabela_mkt = pd.DataFrame(dados_marketing)
display(tabela_mkt)

```

```python
# Gráfico de dispersão
tabela_mkt.plot(kind='scatter', x='Investimento_Anuncios', y='Faturamento',
                     color='orange', title='tabela_Marketing')
```

```python
# Preparação e Treino
from sklearn.linear_model import LinearRegression

X_mkt = tabela_mkt[['Investimento_Anuncios']]
y_mkt = tabela_mkt['Faturamento']

modelo_mkt = LinearRegression()
modelo_mkt.fit(X_mkt, y_mkt)
```

```python
# A pergunta do Diretor: O que acontece se investirmos 8000?

investimento = pd.DataFrame({'Investimento_Anuncios': [8000]})
faturamento_previsto = modelo_mkt.predict(investimento)
print(f"A Estimativa de Retorno no Investimento de 8000 Será de : {faturamento_previsto[0]:.2f}")
```
