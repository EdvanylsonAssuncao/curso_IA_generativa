```python
import pandas as pd
```

```python
arquivo_csv = pd.read_csv('/content/arquivo_csv')

df = pd.DataFrame(arquivo_csv)
```

Atividade 1

```python
maior_salario = df[df['Salário'] == df['Salário'].max()]
print(maior_salario)
```

Atividade 2

```python
menor_salario = df[df['Salário'] == df['Salário'].min()]
print(menor_salario)
```

```python
media = df['Salário'].mean()
print(f"A média da empresa é R${media}")
```

```python
media_matriz = df.groupby('Departamento')['Salário'].mean()
print(f"{round(media_matriz,2)}")
```

```python
qtd_fun_mtz = df['Matriz'].value_counts()
print(qtd_fun_mtz)
print(qtd_fun_mtz.sum())
```
