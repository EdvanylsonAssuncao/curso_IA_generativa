```python
# --- Laços de Repetição (for e while) ---
print("Iniciando a varredura do sistema ...")

# O Laço 'for' vai repetir o código recuado 3 vezes
for numero in range(0, 10, 2):
    print("Analisando número", numero)

print("Varredura concluída!")
```

```python
# --- Importando Bibliotecas ---

import random
import time

print("A IA está pensando em um número ...")
time.sleep(3) # Faz o computador 'dormir' (pausar) por 3 segundos

# Usando a biblioteca random
numero_secreto = random.randint(1, 10)
print("A IA escolheu o número:", numero_secreto)
```

```python
# Converter entrada do input para o número em repetição
numero = input("digite um número: ")
print(numero * 3)

# Converter entrada do input para o número inteiro
numero = int(input("digite um número: "))
print(numero * 3)

# Converter entrada do input para número decimal
numero = float(input("digite um número: "))
print(numero * 3)
```

```python
# Atividade que eu Tentei fazer
# Duelo contra a IA

import random
import time

print("A IA está pensando em um número, voçê tem 3 chances para adivinhar! ...")
time.sleep(3)
numero_secreto = random.randint(1,5)

for numero in range(3):
    numero = int(input("digite um número: "))

    if numero == numero_secreto:
        print("Parabéns voê acertou !!!")
        break
    else:
        print("Infelizmente não foi dessa Vez!")

print(f"O número correto era {numero_secreto}.")





```
