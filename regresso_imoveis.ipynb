```python
# CÉLULA 1: Preparando o terreno
import pandas as pd
from sklearn.linear_model import LinearRegression

# Criando os nossos dados históricos (o passado para ensinar a máquina)
dados = {
    "tamanho_m2": [50, 70, 80, 100, 120],
    "preco": [150000, 210000, 240000, 300000, 360000]
}

tabela = pd.DataFrame(dados)
print("Os nosso dados treino:")
display(tabela)
```

```python
# CÉLULA 2: Gerar um gráfico de dispersão (Scatter Plot)
tabela.plot(kind='scatter', x='tamanho_m2', y='preco', color='blue', title='Relação: Tamanho vs Preço')
```

```python
# CÉLULA 3: Separando X e y
# Para X devemos usar uma tabela (duplos colchetes) e para y usamos uma lista simples
X = tabela[['tamanho_m2']]
y = tabela['preco']
```

```python
# CÉLULA 4: Criando e Treinando a IA
modelo_ia = LinearRegression() # Criamos o "cérebro" vazio

# Treinamos a IA informando os tamanhos (x) e os preços (y)
modelo_ia.fit(X, y)

print("A Intligência Atificial foi treimada com sucesso!")
```

```python
# cÉLULA 5: A hora de verdade (Predição)

# Criamos uma mini-tabela com 1 linha para nova casa, com o mesmo nome da coluna
tamanho_nova_casa = pd.DataFrame({'tamanho_m2': [90]}) # Queramos saber o preço de uma casa de 90m2

# Usamos o comando .predict() para a máquina dar o seu palpite
preco_previsto = modelo_ia.predict(tamanho_nova_casa)

print(f"A IA avaliou a casa em: {preco_previsto[0]:.2f} Reais")
```
