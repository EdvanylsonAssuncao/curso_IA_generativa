```python
# RECONHECIMENTO DE FALA COM ARQUIVO .WAV ---

import speech_recognition as sr

# 1. Inicializar o reconhecedor de voz da IA
reconhecedor = sr.Recognizer()

# 1. Abrir o arquivo de áudio e extrair o texto via IA
nome_arquivo = "detectar.wav"

print(f"[SISTEMA] Abrindo o arquivo '{nome_arquivo}' e preparando os dados...")

# 2. bloco 'with' abre o arquivo de áudio como uma fonte de dados para o Python
with sr.AudioFile(nome_arquivo) as fonte:
    # A IA lê o arquivo e remove pequenos ruídos estáticos de gravação
    reconhecedor.adjust_for_ambient_noise(fonte, duration=0.5)
    dados_audio = reconhecedor.record(fonte)

print("[IA] Enviando o áudio para processamento na rede neural...")

try:
    # Chamar o motor de busca em português (pt-BR)
    texto_transcrito = reconhecedor.recognize_google(dados_audio, language='pt-BR')

    # Padroniza o texto para letras maiúsculas para facilitar a comparação lógica
    comando_final = texto_transcrito.upper()
    print(f"\n[SUCESSO] A IA transcreveu o seguinte texto: \"{comando_final}\"")

# 3. Tratamento de erros caso o áudio esteja mudo ou incompreensível
except sr.UnknownValueError:
    print("\n[ERRO] A IA não conseguiu decifrar os fonemas do áudio. Grave novamente com mais clareza.")
    comando_final = "ERRO"
except Exception as e:
    print(f"\n[ERRO CRÍTICO] Falha na conexão ou arquivo corrompido: {e}")
    comando_final = "ERRO"

# 4. Criar a árvore de decisão baseada na voz
if "DETECTAR" in comando_final:
    print("\n[AÇÃO] >>> Comando Válido! Disparando Rede Neural YOLOv8 (Atividade 2)...")
    # Aqui em um sistema integrado, rodaríamos o código do YOLO

elif "RECONHECER" in comando_final:
    print("\n[AÇÃO] >>> Comando Válido! Disparando Comparação Biométrica Facial (Atividade 3)...")
    # Aqui em um sistema integrado, rodaríamos o código do Face Recognition

else:
    print("\n[ALERTA] >>> Comando não mapeado ou inválido. Sistema mantido em ESPERA.")
```
