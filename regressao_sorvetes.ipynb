```python
# CÉLULA 1: Os Dados e o Gráfico
import pandas as pd

# Od dados históricos da nossa loja
dados_sorvetes = {
    "temperatura_c": [20, 22, 25, 28, 30, 35],
    "sorvetes_vendidos": [50, 65, 80, 100, 120, 180]
}

tabela_sorvetes = pd.DataFrame(dados_sorvetes)
print("O nosso histórico de vendas:")
display(tabela_sorvetes)

# Vamos visualizar a correlação
tabela_sorvetes.plot(kind='scatter', x='temperatura_c', y='sorvetes_vendidos',
                     color='orange', title='Temperatura vs. Vendas de Sorvetes')
```

```python
# CÉLULA 2: O Treino e a Previsão da IA
from sklearn.linear_model import LinearRegression

# 1. Separar a Pergunta (X) da Resposta (y)
X = tabela_sorvetes[['temperatura_c']]
y = tabela_sorvetes['sorvetes_vendidos']

#2. Treinamento o Cérebro da IA
modelo_sorvetes = LinearRegression()
modelo_sorvetes.fit(X, y)

# 3. A Predição: Amanhâ vai fazer uma onda de calor de 38 graus
temperatura_amanha = pd.DataFrame({'temperatura_c': [38]})
predicao = modelo_sorvetes.predict(temperatura_amanha)

print(f"ALERTA DE ESTOQUE: Predição de vendas para amanhã: {int(predicao[0])} sorvetes!")
```
