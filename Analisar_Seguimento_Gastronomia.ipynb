```python
import spacy
import nltk
from nltk.sentiment import SentimentIntensityAnalyzer
!python -m spacy download pt_core_news_lg
nltk.download('vader_lexicon')
nlp = spacy.load("pt_core_news_lg")
termometro_nltk = SentimentIntensityAnalyzer()
```

```python
pesos_extras = {
    # Palavras positivas
    "Acolhimento": 1.0,
    "Nostalgia": 1.5,
    "Sastifação": 3.0,
    "Entusiasmo": 2.0,
    "Conexão": 1.0,
    "Gratidão": 1.0,
    "Felicidade": 1.0,
    # Palavras negativas
    "arrependimento": -2.0,
    "Indiferença": -1.0,
    "ensurdecedor": -4.5,
    "ruim": -4.5,
    "ansiedade": -1.5,
    "barulho": -1.5,
    "liberar": -2.0,
    "estressante": -1.0,
    "Desapontamento": -1.0,
    "invisível": -3.0,
    # Palavras importantes em sentimento
    "Empatia": -0.8,

}
```

```python
frases_clientes = {
    "Mais do que alimentar o corpo, nosso propósito é abraçar a alma com sabores que trazem de volta as melhores memórias da vida.",
    "Ao redor da mesa, o tempo desacelera, as risadas ganham vida e a comida se transforma em amor compartilhável.",
    "Cada ingrediente escolhemos com o coração; cada prato entregamos com a paixão de quem sabe que comer bem é uma das maiores felicidades do mundo.",
    "Nossa cozinha é o nosso lar, e cada cliente que entra por essa porta é recebido não apenas com um prato, mas com um pedaço do nosso carinho.",
    "Acreditamos que a boa gastronomia é uma arte que se sente: um aroma que desperta o desejo, um visual que encanta e um sabor que abraça o paladar.",
    "A fome e a expectativa eram grandes, mas o prato chegou frio e sem sabor; uma decepção que estragou o que era para ser um momento especial.",
    "Mais do que a comida ruim, o que machuca é o atendimento frio; parecer invisível para os garçons faz a gente se sentir um estorvo, e não um cliente.",
    "Olhar para o relógio a cada cinco minutos enquanto os pratos de outras mesas chegam só alimenta a ansiedade e a certeza de que fomos esquecidos pela cozinha.",
    "Pagar caro por uma refeição que entrega menos do que promete deixa um gosto amargo de arrependimento que nenhuma sobremesa consegue apagar.",
    "O barulho ensurdecedor e a pressa da equipe para liberarem a mesa transformaram o jantar em um momento estressante, onde a única vontade era ir embora."
}
```

```python
print(f"{'FEEDBACK ORIGINAL':<70} | {'NOTA':<6} | {'SENTIMENTOS'}")
print("_" * 90)

for texto in frases_clientes:
    texto_minusculo = texto.lower()
    doc = nlp(texto_minusculo)
    palavras_uteis = []
    for token in doc:
        if token.is_punct:
            continue
        if token.is_stop and token.text not in ["Empatia", "ruim", "jamais"]:
            continue
        palavras_uteis.append(token.lemma_)
    print("Palavras úteis:", palavras_uteis)
    frase_limpa = " ".join(palavras_uteis)

    nota = termometro_nltk.polarity_scores(frase_limpa)['compound']
    if nota >= 0.1:
        resultado = " 😊 POSITIVO"
    elif nota <= -0.1:
        resultado = " 😡 NEGATIVO"
    else:
        resultado = " 😐 NEUTRO"

    print(f"{texto:<70} | {nota:6.1f} | {resultado}\n")
```
